1 - Bruno gestiona un pequeño comercio y quiere saber qué producto tuvo el mejor desempeño de ventas el mes pasado. Registró la cantidad vendida de dos productos: manzanas y plátanos. Ahora, necesita escribir un programa que identifique y muestre cuál de ellos tuvo más ventas.

Crea un programa que reciba el número de ventas de los dos productos y muestre un mensaje indicando cuál de ellos vendió más. Si las cantidades son iguales, muestra un mensaje diciendo que hubo un empate.

```python
# Solicitar al usuario la cantidad de manzanas vendidas y convertir a entero
manzanas = int(input("Ingrese la cantidad de manzanas vendidas: "))
# Solicitar al usuario la cantidad de plátanos vendidas y convertir a entero
platanos = int(input("Ingrese la cantidad de plátanos vendidas: "))

# Almacenar las ventas en un diccionario
ventas = {"manzanas": manzanas, "platanos": platanos}

# Comparar las ventas de manzanas y plátanos
if ventas["manzanas"] > ventas["platanos"]:
    # Imprimir mensaje si las manzanas vendieron más
    print((f'El producto que más vendió el mes pasado fueron '
    f'las manzanas con {ventas["manzanas"]} unidades vendidas.'))
elif ventas["platanos"] > ventas["manzanas"]:
    # Imprimir mensaje si los plátanos vendieron más
    print((f'El producto que más vendió el mes pasado fueron '
    f'los plátanos con {ventas["platanos"]} unidades vendidas.'))
else:
    # Imprimir mensaje si las ventas fueron iguales
    print("Las ventas de manzanas y plátanos fueron iguales.")
```

```python
# codigo instructor

# Solicitar al usuario la cantidad de manzanas vendidas y convertir a entero
manzanas = int(input("Digite la cantidad de manzanas vendidas: "))
# Solicitar al usuario la cantidad de plátanos vendidos y convertir a entero
platanos = int(input("Digite la cantidad de plátanos vendidos: "))

# Comparar las ventas de manzanas y plátanos
if manzanas > platanos:
    # Imprimir mensaje si las manzanas tuvieron más ventas
    print("Las manzanas tuvieron más ventas.")
elif platanos > manzanas:
    # Imprimir mensaje si los plátanos tuvieron más ventas
    print("Los plátanos tuvieron más ventas.")
else:
    # Imprimir mensaje si las ventas fueron iguales
    print("Las ventas fueron iguales.")
```

2 - Lucas trabaja en TI y necesita garantizar que la temperatura de una sala de servidores no supere los 25°C. Quiere un programa que reciba la temperatura actual como entrada y, si es necesario, muestre un mensaje de alerta.

```python
# Solicitar al usuario la temperatura actual y convertir a entero
temperatura_actual = int(input("Ingrese la temperatura actual: "))
# Definir el límite de temperatura
temperatura_limite = 25

# Verificar si la temperatura actual supera el límite
if temperatura_actual > temperatura_limite:
  # Imprimir mensaje de alerta si la temperatura es alta
  print((f'¡ALERTA! Temperatura sobre el limite establecido '
  f'de: {temperatura_limite}°C, temperatura actual {temperatura_actual}°C'))
else:
  # Imprimir mensaje si la temperatura está dentro del límite
  print((f'Temperatura dentro de los'
  f'parámetros normales: {temperatura_actual}°C'))
```

```python
#codigo instructor

# Solicitar al usuario la temperatura actual y convertir a flotante
temperatura = float(input("Digite la temperatura actual: "))

# Verificar si la temperatura es mayor a 25
if temperatura > 25:
    # Imprimir mensaje de alerta
    print("¡Alerta! Temperatura por encima del límite permitido.")
else:
    # Imprimir mensaje de seguridad
    print("Temperatura dentro del límite seguro.")
```

3 - Camila está organizando un proyecto y necesita calcular el tiempo total necesario para concluir tres actividades: A, B y C. Sin embargo, si alguna actividad tiene un número de días negativo, el código debe avisar que los valores ingresados son inválidos y no calcular el total.

Escribe un programa que reciba el número de días de tres actividades y muestre el tiempo total del proyecto. Si algún valor es negativo, muestra un mensaje informando el error.

```python
#codigo basico

# Solicitar al usuario la duración de la actividad A y convertir a entero
proyecto_a = int(input("Ingrese el tiempo necesario para conlcuir la actividad A: "))
# Solicitar al usuario la duración de la actividad B y convertir a entero
proyecto_b = int(input("Ingrese el tiempo necesario para conlcuir la actividad B: "))
# Solicitar al usuario la duración de la actividad C y convertir a entero
proyecto_c = int(input("Ingrese el tiempo necesario para conlcuir la actividad C: "))

# Verificar si todas las duraciones son no negativas
if proyecto_a > 0 and proyecto_b > 0 and proyecto_c > 0:
  # Calcular y imprimir la duración total si los valores son válidos
  print(f'La duracion total del proyecto es {sum([proyecto_a, proyecto_b, proyecto_c])} días')
else:
 # Imprimir mensaje de error si alguna duración es negativa o cero
 print(f'La duracion no puede ser negativa ni cero')
```

```python
# mejora, error al primer dato no valido.

# Inicializar lista para almacenar duraciones
duraciones = []
# Lista de nombres de actividades
actividades = ['A', 'B', 'C']
# Bandera para verificar si todos los valores son válidos
valores_validos = True

# Iterar sobre cada actividad
for actividad in actividades:
    # Solicitar al usuario la duración de la actividad actual
    tiempo = int(input(f"Ingrese el tiempo necesario para concluir la actividad {actividad}: "))
    # Verificar si el tiempo es negativo o cero
    if tiempo <= 0:
        # Imprimir mensaje de error
        print('La duracion no puede ser negativa ni cero')
        # Establecer la bandera a False si hay un valor inválido
        valores_validos = False
        break # Salir del bucle si se ingresa un valor inválido
    # Agregar la duración válida a la lista
    duraciones.append(tiempo)

# Si todos los valores fueron válidos
if valores_validos:
    # Calcular y imprimir la duración total del proyecto
    print(f'La duracion total del proyecto es {sum(duraciones)} días')
```

```python
# Mejora mas robusta

def obtener_duracion_actividad(actividad):
    """Solicita y valida la duración de una actividad."""
    while True:
        try:
            # Solicitar la duración al usuario y convertir a entero
            tiempo = int(input(f"Ingrese el tiempo necesario para concluir la actividad {actividad}: "))
            # Verificar si el tiempo es positivo
            if tiempo > 0:
                return tiempo # Devolver el valor si es válido
            else:
                # Imprimir mensaje si la duración no es positiva
                print("La duración no puede ser negativa ni cero. Intente de nuevo.")
        except ValueError:
            # Manejar el error si la entrada no es un número entero
            print("Entrada inválida. Por favor, ingrese un número entero. Intente de nuevo.")

# Inicializar lista para almacenar duraciones
duraciones = []
# Lista de nombres de actividades
actividades = ['A', 'B', 'C'] # Puedes agregar más actividades aquí

# Imprimir instrucciones para el usuario
print("Ingrese las duraciones de las actividades. La duración debe ser un número entero positivo.")

# Iterar sobre cada actividad y obtener su duración validada
for actividad in actividades:
    duracion = obtener_duracion_actividad(actividad)
    duraciones.append(duracion)

# Calcular la duración total del proyecto
total_dias = sum(duraciones)
# Imprimir la duración total
print(f'\nLa duración total del proyecto es {total_dias} días')
```
