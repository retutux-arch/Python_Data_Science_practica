Crea un programa que reciba el número de ventas de los dos productos y muestre un mensaje indicando cuál de ellos vendió más. Si las cantidades son iguales, muestra un mensaje diciendo que hubo un empate.

```python
# Solicita al usuario la cantidad de manzanas vendidas y la convierte a entero
manzanas = int(input("Ingrese la cantidad de manzanas vendidas: "))
# Solicita al usuario la cantidad de plátanos vendidas y la convierte a entero
platanos = int(input("Ingrese la cantidad de plátanos vendidas: "))

# Almacena las ventas en un diccionario
ventas = {"manzanas": manzanas, "platanos": platanos}

# Compara las ventas de manzanas y plátanos
if ventas["manzanas"] > ventas["platanos"]:
    # Imprime un mensaje si las manzanas vendieron más
    print((f'El producto que más vendió el mes pasado fue '
    f'manzanas con {ventas["manzanas"]} unidades vendidas.'))
elif ventas["platanos"] > ventas["manzanas"]:
    # Imprime un mensaje si los plátanos vendieron más
    print((f'El producto que más vendió el mes pasado fueron '
    f'plátanos con {ventas["platanos"]} unidades vendidas.'))
else:
    # Imprime un mensaje si las ventas fueron iguales
    print("Las ventas de manzanas y plátanos fueron iguales.")
```

3 - Camila está organizando un proyecto y necesita calcular el tiempo total necesario para concluir tres actividades: A, B y C. Sin embargo, si alguna actividad tiene un número de días negativo, el código debe avisar que los valores ingresados son inválidos y no calcular el total.

Escribe un programa que reciba el número de días de tres actividades y muestre el tiempo total del proyecto. Si algún valor es negativo, muestra un mensaje informando el error.

```python
# codigo básico

# Solicita al usuario la duración de la actividad A y la convierte a entero
proyecto_a = int(input("Ingrese el tiempo necesario para conlcuir la actividad A: "))
# Solicita al usuario la duración de la actividad B y la convierte a entero
proyecto_b = int(input("Ingrese el tiempo necesario para conlcuir la actividad B: "))
# Solicita al usuario la duración de la actividad C y la convierte a entero
proyecto_c = int(input("Ingrese el tiempo necesario para conlcuir la actividad C: "))

# Verifica si todas las duraciones no son negativas
if proyecto_a > 0 and proyecto_b > 0 and proyecto_c > 0:
  # Calcula e imprime la duración total si los valores son válidos
  print(f'La duración total del proyecto es {sum([proyecto_a, proyecto_b, proyecto_c])} días')
else:
 # Imprime un mensaje de error si alguna duración es negativa o cero
 print(f'La duración no puede ser negativa o cero')
```

```python
# mejora, error en el primer dato inválido.

# Inicializa una lista para almacenar las duraciones
duraciones = []
# Lista de nombres de actividades
actividades = ['A', 'B', 'C']
# Bandera para verificar si todos los valores son válidos
valores_validos = True

# Itera sobre cada actividad
for actividad in actividades:
    # Solicita al usuario la duración de la actividad actual
    tiempo = int(input(f"Ingrese el tiempo necesario para concluir la actividad {actividad}: "))
    # Verifica si el tiempo es negativo o cero
    if tiempo <= 0:
        # Imprime un mensaje de error
        print('La duración no puede ser negativa o cero')
        # Establece la bandera a False si hay un valor inválido
        valores_validos = False
        break # Sale del bucle si se ingresa un valor inválido
    # Agrega la duración válida a la lista
    duraciones.append(tiempo)

# Si todos los valores fueron válidos
if valores_validos:
    # Calcula e imprime la duración total del proyecto
    print(f'La duración total del proyecto es {sum(duraciones)} días')
```

```python
# Mejora más robusta

def obtener_duracion_actividad(actividad):
    """Solicita y valida la duración de una actividad."""
    while True:
        try:
            # Solicita la duración al usuario y la convierte a entero
            tiempo = int(input(f"Ingrese el tiempo necesario para concluir la actividad {actividad}: "))
            # Verifica si el tiempo es positivo
            if tiempo > 0:
                return tiempo # Retorna el valor si es válido
            else:
                # Imprime un mensaje si la duración no es positiva
                print("La duración no puede ser negativa o cero. Intente de nuevo.")
        except ValueError:
            # Maneja el error si la entrada no es un entero
            print("Entrada inválida. Por favor, ingrese un número entero. Intente de nuevo.")

# Inicializa una lista para almacenar las duraciones
duraciones = []
# Lista de nombres de actividades
actividades = ['A', 'B', 'C'] # Puedes agregar más actividades aquí

# Imprime instrucciones para el usuario
print("Ingrese las duraciones de las actividades. La duración debe ser un número entero positivo.")

# Itera sobre cada actividad y obtiene su duración validada
for actividad in actividades:
    duracion = obtener_duracion_actividad(actividad)
    duraciones.append(duracion)

# Calcula la duración total del proyecto
total_dias = sum(duraciones)
# Imprime la duración total
print(f'\nLa duración total del proyecto es {total_dias} días')
```

```python
# Código del instructor

actividad_A = int(input("Informe os dias para a atividade A: ")) # Keeping this message as it was in the original instructor code
actividad_B = int(input("Informe os dias para a atividade B: ")) # Keeping this message as it was in the original instructor code
actividad_C = int(input("Informe os dias para a atividade C: ")) # Keeping this message as it was in the original instructor code

if actividad_A < 0 or actividad_B < 0 or actividad_C < 0:
    print("Error: Os dias não podem ser negativos.") # Keeping this message as it was in the original instructor code
else:
    tiempo_total = actividad_A + actividad_B + actividad_C
    print(f"El tiempo total del proyecto es de {tiempo_total} días.") # Keeping this message as it was in the original instructor code
```

4 - Anna Júlia está creando un sistema para calcular el Índice de Masa Corporal (IMC) y proporcionar recomendaciones básicas. El programa debe recibir el peso y la altura de una persona y mostrar el valor del IMC, además de indicar si está por debajo del peso, con peso normal o por encima del peso. Crea un programa que reciba el peso (en kg) y la altura (en metros) y calcule el IMC usando la fórmula: IMC = peso / (altura ** 2)Luego, muestra el valor del IMC y un mensaje indicando si está por debajo del peso (IMC < 18.5), peso normal (18.5 <= IMC < 25) o por encima del peso (IMC >= 25).

```python
peso_usuario = float(input("Ingrese su peso en KG:  "))
altura_usuario = float(input("Ingrese su altura en metros:  "))

calculo_imc = peso_usuario / (altura_usuario **2)
print(f'Su IMC es : {calculo_imc:.2f}')

if calculo_imc < 18.5 :
  print("Su peso está por debajo de los parámetros recomendados")
elif 18.5 <= calculo_imc < 25 : # Condición ajustada para peso normal
  print('Su peso es normal.')
else: # Esto cubre IMC >= 25
  print ('Su peso está por encima de los parámetros recomendados')
```

```python
# codigo instructor

# Solicita al usuario su peso en kilogramos y lo convierte a flotante
peso = float(input("Digite su peso (kg): "))
# Solicita al usuario su altura en metros y lo convierte a flotante
altura = float(input("Digite su altura (m): "))

# Calcula el Índice de Masa Corporal (IMC)
imc = peso / (altura ** 2)
# Imprime el valor del IMC redondeado a dos decimales
print(f"Su IMC es: {imc:.2f}")

# Compara el IMC con los rangos definidos para determinar el estado de peso
if imc < 18.5:
    # Si el IMC es menor a 18.5, imprime que está por debajo del peso
    print("Estás por debajo del peso.")
elif imc < 25:
    # Si el IMC es menor a 25 (pero no menor a 18.5, debido al if anterior), imprime que tiene un peso normal
    print("Tienes un peso normal.")
else:
    # Si el IMC no es menor a 25, imprime que está por encima del peso
    print("Estás por encima del peso.")
```

5 - Laura está desarrollando un sistema para saber si una persona tiene derecho a recibir un beneficio social. Para eso, la persona debe cumplir las siguientes condiciones:

*   Tener al menos un hijo o hija.

Crea un programa que reciba los ingresos mensuales y la cantidad de hijos de una persona, y diga si tiene derecho al beneficio.

```python
# Sistema poco ético.

# Solicita al usuario sus ingresos mensuales y los convierte a entero
ingreso = int(input("Digite sus ingresos:  "))
# Solicita al usuario la cantidad de hijos y la convierte a entero
hijos = int(input("Digite la cantidad de hijos:  "))

# Verifica si el ingreso es menor o igual a 2000 Y si la cantidad de hijos es mayor a 0
if ingreso <= 2000 and hijos > 0 :
  # Si ambas condiciones son verdaderas, imprime que tiene derecho al beneficio
  print("Tiene derecho al beneficio social")
else:
  # Si alguna de las condiciones es falsa, imprime que no tiene derecho al beneficio
  print("No tiene derecho al beneficio social")
```

```python
#Éticamente correcto

# Solicita al usuario sus ingresos mensuales, cantidad de hijos y si es un hogar monoparental.
ingreso = float(input("Ingrese sus ingresos mensuales: "))
hijos = int(input("Ingrese la cantidad de hijos: "))
es_monoparental = input("¿Es un hogar monoparental? (sí/no): ").lower()

# Define el límite base de ingresos y los incrementos
limite_base = 1500
incremento_por_hijo = 500
incremento_monoparental = 700

# Calcula el límite de ingresos ajustado
limite_ajustado = limite_base + (incremento_por_hijo * hijos)

# Agrega el incremento para hogares monoparentales si aplica
if es_monoparental == 'sí':
    limite_ajustado += incremento_monoparental

# Implementa la lógica condicional para verificar la elegibilidad
if hijos > 0 and ingreso <= limite_ajustado:
    print("Tiene derecho al beneficio social.")
else:
    print("No tiene derecho al beneficio social.")
```
