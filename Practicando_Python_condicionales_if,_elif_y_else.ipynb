Crea un programa que reciba el número de ventas de los dos productos y muestre un mensaje indicando cuál de ellos vendió más. Si las cantidades son iguales, muestra un mensaje diciendo que hubo un empate.

```python
# Solicita al usuario la cantidad de manzanas vendidas y la convierte a entero
manzanas = int(input("Ingrese la cantidad de manzanas vendidas: "))
# Solicita al usuario la cantidad de plátanos vendidas y la convierte a entero
platanos = int(input("Ingrese la cantidad de plátanos vendidas: "))

# Almacena las ventas en un diccionario
ventas = {"manzanas": manzanas, "platanos": platanos}

# Compara las ventas de manzanas y plátanos
if ventas["manzanas"] > ventas["platanos"]:
    # Imprime un mensaje si las manzanas vendieron más
    print((f'El producto que más vendió el mes pasado fue '
    f'manzanas con {ventas["manzanas"]} unidades vendidas.'))
elif ventas["platanos"] > ventas["manzanas"]:
    # Imprime un mensaje si los plátanos vendieron más
    print((f'El producto que más vendió el mes pasado fueron '
    f'plátanos con {ventas["platanos"]} unidades vendidas.'))
else:
    # Imprime un mensaje si las ventas fueron iguales
    print("Las ventas de manzanas y plátanos fueron iguales.")
```

El programa debe recibir el peso y la altura de una persona y mostrar el valor del IMC, además de indicar si está por debajo del peso, con peso normal o por encima del peso. Crea un programa que reciba el peso (en kg) y la altura (en metros) y calcule el IMC usando la fórmula: IMC = peso / (altura ** 2)Luego, muestra el valor del IMC y un mensaje indicando si está por debajo del peso (IMC < 18.5), peso normal (18.5 <= IMC < 25) o por encima del peso (IMC >= 25).

```python
# Solicita al usuario su peso en kilogramos y lo convierte a flotante
peso_usuario = float(input("Ingrese su peso en KG:  "))
# Solicita al usuario su altura en metros y lo convierte a flotante
altura_usuario = float(input("Ingrese su altura en metros:  "))

# Calcula el Índice de Masa Corporal (IMC)
calculo_imc = peso_usuario / (altura_usuario **2)
# Imprime el valor del IMC redondeado a dos decimales
print(f'Su IMC es : {calculo_imc:.2f}')

# Compara el IMC con los rangos definidos para determinar el estado de peso
if calculo_imc < 18.5 :
  print("Su peso está por debajo de los parámetros recomendados")
elif 18.5 <= calculo_imc < 25 :
  print('Su peso es normal.')
else:
  print ('Su peso está por encima de los parámetros recomendados')
```

```python
# codigo instructor

# Solicita al usuario su peso en kilogramos y lo convierte a flotante
peso = float(input("Digite su peso (kg): "))
# Solicita al usuario su altura en metros y la convierte a flotante
altura = float(input("Digite su altura (m): "))

# Calcula el Índice de Masa Corporal (IMC)
imc = peso / (altura ** 2)
# Imprime el valor del IMC redondeado a dos decimales
print(f"Su IMC es: {imc:.2f}")

# Compara el IMC con los rangos definidos para determinar el estado de peso
if imc < 18.5:
    # Si el IMC es menor a 18.5, imprime que está por debajo del peso
    print("Estás por debajo del peso.")
elif imc < 25:
    # Si el IMC es menor a 25 (pero no menor a 18.5, debido al if anterior), imprime que tiene un peso normal
    print("Tienes un peso normal.")
else:
    # Si el IMC no es menor a 25, imprime que está por encima del peso
    print("Estás por encima del peso.")
```

9 - Una empresa evalúa su trimestre con base en:

* Ingresos totales
* Gastos totales
* Número de nuevos clientes

Clasificación:

* Si ingresos - gastos > $10,000 y más de 50 nuevos clientes → "Trimestre Excelente"

* Si ingresos - gastos > $5,000 y al menos 20 clientes → "Trimestre Bueno"
* Si ingresos - gastos > 0 → "Trimestre Regular"
* Si ingresos - gastos ≤ 0 → "Trimestre Deficitario"

```python
# Este código clasifica el trimestre de una empresa basado en ingresos, gastos y nuevos clientes.

# Solicita al usuario el total de ingresos y lo convierte a entero
total_ingresos = int(input("Ingrese el total de ingresos: "))
# Solicita al usuario el total de gastos y lo convierte a entero
total_gastos = int(input("Ingrese el total de gastos: "))
# Solicita al usuario el número de nuevos clientes y lo convierte a entero
nuevos_clientes = int(input("Ingrese el número de nuevos clientes: "))

# Calcula la diferencia entre ingresos y gastos
diferencia = total_ingresos - total_gastos

# Clasifica el trimestre según las reglas
if diferencia > 10000 and nuevos_clientes >= 50 :
  print("Trimestre Excelente")
elif diferencia > 5000 and nuevos_clientes >= 20 :
  print("Trimestre Bueno")
elif diferencia > 0 :
  print("Trimestre Regular")
else:
  print("Trimestre Deficitario")
```

```python
# codigo instructor

# Solicita al usuario los ingresos del trimestre y los convierte a flotante
ingresos = float(input("Ingrese los ingresos del trimestre: "))
# Solicita al usuario los gastos del trimestre y los convierte a flotante
gastos = float(input("Ingrese los gastos del trimestre: "))
# Solicita al usuario el número de nuevos clientes y lo convierte a entero
clientes_nuevos = int(input("Ingrese el número de nuevos clientes: "))

# Calcula el beneficio restando los gastos de los ingresos
beneficio = ingresos - gastos

# Verifica las condiciones para clasificar el trimestre
if beneficio > 10000 and clientes_nuevos > 50:
    # Si el beneficio es mayor a 10000 y los clientes nuevos son más de 50, es un trimestre Excelente
    print("Trimestre Excelente")
elif beneficio > 5000 and clientes_nuevos >= 20:
    # Si el beneficio es mayor a 5000 y los clientes nuevos son al menos 20 (y no cumple la condición anterior), es un trimestre Bueno
    print("Trimestre Bueno")
elif beneficio > 0:
    # Si el beneficio es mayor a 0 (y no cumple las condiciones anteriores), es un trimestre Regular
    print("Trimestre Regular")
else:
    # Si el beneficio es 0 o menor, es un trimestre Deficitario
    print("Trimestre Deficitario")
```
