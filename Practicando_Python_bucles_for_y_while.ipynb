# ¿Qué es un bucle infinito?

André está probando una nueva función en el backend de Buscante que procesa datos en un bucle. Durante las pruebas, se dio cuenta de que el sistema dejó de responder y sospecha que el problema está en un bucle infinito.



```
contador = 0

while contador < 10:
    print("Procesando datos...")
```



¿Cuál es el problema del código de André y cómo resolverlo?

````python
contador = 0

while contador < 10:
    print(f"Procesando datos...{contador}")
    contador += 1 # el codigo no poseia un codigo para terminar el bucle.
````

#Organizando tu portafolio

Ana está desarrollando su portafolio para exhibir los proyectos de Python que ha completado. Ella organizó una lista con el nombre de cada proyecto, pero se dio cuenta de que algunos elementos pueden estar ausentes, apareciendo como None:

proyectos = ["sitio web", "juego", "análisis de datos", None, "aplicativo móvil"]

Crea un programa que ayude a Ana a recorrer la lista de proyectos y muestre los nombres de los proyectos válidos. Si encuentra un elemento None, el programa debe mostrar el mensaje: "Proyecto ausente".

**Salida esperada:**

* sitio web
* juego
* análisis de datos
* Proyecto ausente
* aplicativo móvil

````python
proyectos = ["sitio web", "juego", "análisis de datos", None, "aplicativo móvil"]

for proyecto in proyectos :
  if proyecto is not None: #si no es NONE imrpime proyecto
    print(proyecto)
  else:
    print("Proyecto ausente")# si es NONE, lo reemplaza por "Proyecto ausente"
````

# Entendiendo el uso del break


José está desarrollando una funcionalidad en el sistema de Buscante para interrumpir la búsqueda tan pronto como se encuentre un libro específico. La lista de libros ya registrados en el sistema es la siguiente:

`libros = ["1984", "Cien años de soledad", "El Principito", "El Hobbit", "Orgullo y Prejuicio"]`

Ayuda a José a crear un programa que recorra la lista y muestre el mensaje "Libro encontrado: <nombre del libro>" tan pronto como se encuentre el libro "El Hobbit". Después de encontrar el libro, el programa debe detener inmediatamente la búsqueda, sin verificar los libros restantes.

Salida esperada:

   libro encontrado : El Hobbit
   

````python
libros = ["1984", "Cien años de soledad", "El Principito", "El Hobbit", "Orgullo y Prejuicio"]

for libro in libros:
  if libro == "El Hobbit":
    print(f'Libro encontrado: {libro}')
    break

print("Búsqueda terminada.")
````

````python
# codigo verificacion
''' Sin la línea break del código, el bucle for seguirá iterando a través
 de todos los elementos de la lista libros, incluso después de haber encontrado y
 reportado que "El Hobbit" fue encontrado.  '''

libros = ["1984", "Cien años de soledad", "El Principito", "El Hobbit", "Orgullo y Prejuicio"]

print("Iniciando búsqueda...")
for libro in libros:
  print(f"Verificando: {libro}") # Añadimos este print para ver cada iteración
  if libro == "El Hobbit":
    print(f'Libro encontrado: {libro}')
    break # Aquí, el break detendrá el bucle inmediatamente

print("Búsqueda terminada.") # Este print nos dirá si el bucle completó o se detuvo
````

# Control de inventario


Estás desarrollando un sistema de control de inventario para Buscante. Uno de los requisitos es verificar la cantidad de ejemplares de un libro en inventario y continuar vendiendo hasta que el inventario se agote. Siempre que se realiza una venta, el sistema debe informar al usuario y actualizar la cantidad disponible.

Crea un programa que simule las ventas de un libro con el inventario inicial de 5 ejemplares. El programa debe mostrar el mensaje "¡Venta realizada! Inventario restante: <cantidad>" con cada venta y, al final, mostrar el mensaje "Inventario agotado".

**Salida esperada:**
  
  "¡Venta realizada! Inventario restante: 5

  "¡Venta realizada! Inventario restante: 4

  "¡Venta realizada! Inventario restante: 3

  "¡Venta realizada! Inventario restante: 2

  "¡Venta realizada! Inventario restante: 1

  Intenraio agotado


````python
libros = ["1984", "Cien años de soledad", "El Principito", "El Hobbit", "Orgullo y Prejuicio",]

contador_libros = len(libros)
total_inventario = contador_libros

print(f'¡Venta realizada! Inventario restante: {len(libros)}')


for i in range(total_inventario):
  total_inventario -= 1
  if total_inventario > 0:
    print(f'¡Venta realizada! Inventario restante: {total_inventario}')


print("Inventario agotado")
````

````python
libros = ["1984", "Cien años de soledad", "El Principito", "El Hobbit", "Orgullo y Prejuicio"]

print(f"Inventario inicial ({len(libros)} libros):", libros)

print("\nIniciando simulación de ventas (ingresa 'salir' para terminar):")

# Crear una lista de libros en minúsculas para facilitar la comparación
libros_lower = [libro.lower() for libro in libros]

while libros: # El bucle continúa mientras la lista de libros no esté vacía
  print(f"\nInventario actual ({len(libros)} libros):", libros)
  libro_a_vender = input("Ingresa el nombre del libro a vender: ").lower() # Convertir la entrada del usuario a minúsculas

  if libro_a_vender == 'salir':
    print("Simulación de ventas terminada por el usuario.")
    break # Salir del bucle si el usuario ingresa 'salir'

  if libro_a_vender in libros_lower: # Comparar con la lista de libros en minúsculas
    # Encontrar el nombre original del libro para eliminarlo de la lista original
    index_to_remove = libros_lower.index(libro_a_vender)
    libro_vendido = libros.pop(index_to_remove)
    libros_lower.pop(index_to_remove) # Eliminar también de la lista en minúsculas

    print(f'¡Venta realizada de "{libro_vendido}"!')
    print(f"Inventario restante: {len(libros)} libros") # Ya estaba len() aquí
  else:
    print(f'"{libro_a_vender}" no se encuentra en el inventario. Intenta de nuevo.')

if not libros: # Verificar si el bucle terminó porque el inventario se agotó
  print("\nInventario agotado")
````
