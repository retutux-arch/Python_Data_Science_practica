```python
# Opcion mas robusta y actual.
'''con Generator obtienes una pseudoaleatoriedad de mayor calidad, lo que en
la práctica lo hace más parecido a la aleatoriedad verdadera en términos
estadísticos y de robustez, aunque sigue siendo pseudoaleatorio.'''


lista = [8, 12, 54, 23, 43, 1, 90, 87, 105, 77]

# Crear una instancia de Generator (el generador de números aleatorios recomendado)
rng = np.random.default_rng()

# Usar el método choice de esa instancia
numero_aleatorio = rng.choice(lista)

print(f"El número elegido al azar de la lista es: {numero_aleatorio}")
```

```python
# Codigo instructor

import random

lista = [8, 12, 54, 23, 43, 1, 90, 87, 105, 77]
numero_aleatorio = random.choice(lista)
print(numero_aleatorio)
```

### Ejercicio 5:
Crea un programa que solicite a la persona usuaria ingresar dos números enteros y calcule la potencia del primer número elevado al segundo.

```python
# Este código utiliza el operador aritmético de Python '**' para calcular la potencia.
# Es la forma más directa y comúnmente utilizada en Python.

primer_numero = int(input("Ingrese el primer número: "))
segundo_numero = int(input("Ingrese el segundo número: "))

resultado = primer_numero ** segundo_numero

print(f"El resultado de {primer_numero} elevado a la {segundo_numero} es: {resultado}")
```

```python
#version Martillo hidraulico for clavo

import numpy as np

# Este código utiliza la función 'power()' de la biblioteca numpy.
# Es útil especialmente cuando se trabaja con arrays de NumPy, aplicando la operación elemento por elemento.

primer_numero = int(input("Ingrese el primer número: "))
segundo_numero = int(input("Ingrese el segundo número: "))

resultado = np.power(primer_numero, segundo_numero)

print(f"El resultado de {primer_numero} elevado a la {segundo_numero} es: {resultado}")
```

```python
# Este código utiliza la función 'pow()' que es una función incorporada (built-in) de Python.
# Es equivalente al operador '**' y puede tomar un tercer argumento para el módulo.

primer_numero = int(input("Ingrese el primer número: "))
segundo_numero = int(input("Ingrese el segundo número: "))

resultado = pow(primer_numero, segundo_numero)

print(f"El resultado de {primer_numero} elevado a la {segundo_numero} es: {resultado}")
```

```python
#codigo instructor

''' La bioblioteca de math funciona en este caso, pero siempre entrega un resultado flotante,
dado que el ejercicio solicita un entero, no es la mejor practica.'''

import math
num1 = int(input("Ingrese el primer número entero: "))
num2 = int(input("Ingrese el segundo número entero: "))
resultado = math.pow(num1, num2)
print(f"La potencia de {num1} elevado a {num2} es {resultado}")
```

### Ejercicio 7 (Aplicando a proyectos):
Has recibido una solicitud para generar números de token para acceder a la aplicación de una empresa. El token debe ser par y variar de 1000 a 9998. Escribe un código que solicite el nombre de la persona usuaria y muestre un mensaje junto a este token generado aleatoriamente:
`print(f"Hola, {nombre_usuario}, tu token de acceso es {token_generado} ¡Bienvenido/a!")`

```python
import random

usuario = input("Ingrese su nombre de usuario: ")

# random.randrange(inicio, fin_exclusivo, paso)
token = random.randrange(1000, 9999, 2) # Genera un número par entre 1000 y 9998 (9999 es exclusivo)

print(f"Hola, {usuario}, tu token de acceso es {token} ¡Bienvenido/a!")
```

```python
#version Martillo hidraulico for clavo

import numpy as np

usuario = input("Ingrese su nombre de usuario: ")

token_par = np.arange(1000,9999,2)

token = np.random.choice(token_par)

print(f"Hola, {usuario}, tu token de acceso es {token} ¡Bienvenido/a!")
```
